```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv('/content/Consumer_Complaints_train.csv')
test=pd.read_csv('/content/Consumer_Complaints_test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
#Do missing value analysis and drop columns where more than 25% of data are missing
missing_values_train = train.isnull().sum()
missing_values_test = test.isnull().sum()
print("Missing values per column in train data:\n", missing_values_train)
print("\nMissing values per column in test data:\n", missing_values_test)
```

```python
# Calculate percentage of missing values
missing_percentage_train = train.isnull().mean() * 100
missing_percentage_test = test.isnull().mean() * 100
print("\nPercentage of missing values per column in train data:\n", missing_percentage_train)
print("\nPercentage of missing values per column in test data:\n", missing_percentage_test)
```

```python
# Drop columns with more than 25% missing values
threshold = 25
cleaned_train_data= train.loc[:, train.isnull().mean() * 100 < threshold]
cleaned_test_data= test.loc[:, test.isnull().mean() * 100 < threshold]
print("\nDataFrame after dropping columns with more than 25% missing values:\n")
cleaned_test_data.head()
```

```python
cleaned_train_data.head()
```

```python
train_c=cleaned_train_data.copy()
test_c=cleaned_test_data.copy()
```

```python
# Convert 'Date_Received' to datetime format
train_c['Date received'] = pd.to_datetime(train_c['Date received'])
test_c['Date received'] = pd.to_datetime(test_c['Date received'])
# Extract day, month, and year
train_c['Day'] = train_c['Date received'].dt.day
train_c['Month'] = train_c['Date received'].dt.month
train_c['Year'] = train_c['Date received'].dt.year
test_c['Day'] = test_c['Date received'].dt.day
test_c['Month'] = test_c['Date received'].dt.month
test_c['Year'] = test_c['Date received'].dt.year
```

```python
# Calculate the number of days the complaint was with the company
train_c['Date sent to company'] = pd.to_datetime(train_c['Date sent to company'])
train_c['Days_Held'] = (train_c['Date sent to company'] - train_c['Date received']).dt.days
test_c['Date sent to company'] = pd.to_datetime(test_c['Date sent to company'])
test_c['Days_Held'] = (test_c['Date sent to company'] - test_c['Date received']).dt.days
```

```python
#Imputing Null value in “State” by Mode
train_c['State'].fillna(train_c['State'].mode()[0], inplace=True)
test_c['State'].fillna(test_c['State'].mode()[0], inplace=True)
```

```python
# with the help of the days we calculated above, create a new field 'Week_Received' where we calculate the week based on the day of receiving.
train_c['Week_Received'] = train_c['Day'].apply(lambda x: (x-1)//7 + 1)
test_c['Week_Received'] = test_c['Day'].apply(lambda x: (x-1)//7 + 1)
```

```python
#store data of disputed people into the “disputed_cons” variable for future tasks
disputed_cons=train_c[train_c['Company response to consumer']=='Disputed']
```

```python
train_c.to_csv('train_c.csv', index=False)
test_c.to_csv('test_c.csv', index=False)
```
